```python
plt.style.use ('fivethirtyeight')
slices= [5.3,4.5,6.3]
labels = ['Nigeria', 'South Sudan', 'Chad']
colors = ['#dda0dd', '#a0ddd2', '#b4a0dd']
explode = [0,0,0.1]


 #['#008fd5', '#e5ae37', '#6d984f']
plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, autopct= '%1.1f%%',colors=colors, wedgeprops={'edgecolor':'black'})

plt.title('Fertility rate, total (births per woman)')
plt.tight_layout()
plt.show()
```
